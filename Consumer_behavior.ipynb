```python
df1['Review Rating']=df1.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df1['Review Rating'].describe()
```

```python
df1.columns=df1.columns.str.lower()
df1.columns=df1.columns.str.replace(' ','_')
df1.columns
```

```python
df1.head()
```

```python
label=['Young','Adult','Middle_aged','Senior']
df1['age_segment']=pd.qcut(df1['age'],q=4,labels=label)
```

```python
df1.loc[df1['age_segment']=='Adult',['age','age_segment']]
```

```python
df1['age_seg']=pd.cut(df1['age'],bins=4,labels=label)
df1
```

```python
def section(age):
    if age > 55:
        return 'Senior'
    elif age > 40:
        return 'Middle_age'
    elif age > 20:
        return 'Adult'
    else:
        return 'Youth'

df1['section']=df1['age'].apply(section)
```

```python
df1=df1.drop('section',axis=1)
```

```python
df1
```

```python
df1['frequency_of_purchases'].value_counts()
```

```python
frequency_mapping={
    'Every 3 Months':90,
    'Annually':365,
    'Quarterly':90,
    'Monthly':30,
    'Bi-Weekly':14,
    'Fortnightly':14,
    'Weekly':7
}
df1['purchase_frequency_days']=df1['frequency_of_purchases'].map(frequency_mapping)
```
